```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

BASE = "https://www.mayoclinic.org"
INDEX_URL = "https://www.mayoclinic.org/diseases-conditions/index"

headers = {
    "User-Agent": "Mozilla/5.0"
}

all_data = []

session = requests.Session()
session.headers.update(headers)

# Step 1: Loop A–Z
for letter in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
    print(f"\nProcessing letter {letter} ...")

    index_page = session.get(INDEX_URL, params={"letter": letter})
    soup = BeautifulSoup(index_page.text, "html.parser")

    # Step 2: Get all disease links for this letter
    tags = soup.find_all(
        "a",
        class_="cmp-anchor--plain cmp-button cmp-button__link cmp-results-with-primary-name__see-link"
    )

    disease_links = [tag["href"] for tag in tags]

    print(f"Found {len(disease_links)} diseases for letter {letter}")

    # Step 3: Visit each disease page
    for link in disease_links:
        full_link = BASE + link if link.startswith("/") else link

        try:
            disease_page = session.get(full_link)
            disease_soup = BeautifulSoup(disease_page.text, "html.parser")

            # Disease name
            disease_name = disease_soup.find("h1").get_text(strip=True)

            wanted_sections = ["Overview", "Symptoms", "Causes", "Risk factors"]
            section_data = {sec: "" for sec in wanted_sections}

            # Extract sections
            for h2 in disease_soup.find_all("h2"):
                title = h2.get_text(strip=True)

                if title in wanted_sections:
                    content = ""
                    for sibling in h2.find_next_siblings():
                        if sibling.name == "h2":
                            break
                        content += sibling.get_text(" ", strip=True) + " "

                    section_data[title] = content.strip()

            all_data.append({
                "Disease": disease_name,
                "Link": full_link,
                "Overview": section_data["Overview"],
                "Symptoms": section_data["Symptoms"],
                "Causes": section_data["Causes"],
                "Risk Factors": section_data["Risk factors"]
            })

            print("Saved:", disease_name)

            time.sleep(1)  # polite delay

        except Exception as e:
            print("Error:", e)

# Step 4: Save to CSV
df = pd.DataFrame(all_data)
df.to_csv("data/mayo_clinic_all_diseases.csv", index=False)

print("\nDONE. File saved as mayo_clinic_all_diseases.csv")
```

```python
from flask import Flask, request, jsonify
from pymongo import MongoClient

app = Flask(__name__)

client = MongoClient('mongodb://localhost:27017/')
db = client['my_database']
collection = db['users']

def seed_database():
    if collection.count_documents({}) == 0:
        print("Database is empty. Inserting initial seed data...")
        
        seed_data = [
            {"name": "John Doe", "email": "john@example.com", "age": 28},
            {"name": "Jane Smith", "email": "jane@example.com", "age": 34},
            {"name": "Bob Martin", "email": "bob@example.com", "age": 45}
        ]
        
        collection.insert_many(seed_data)
        print("Seed data inserted successfully.")
    else:
        print("Database already contains data. Skipping seeding.")

@app.route('/add_user', methods=['POST'])
def add_user():
    data = request.json
    if not data or 'name' not in data:
        return jsonify({"error": "Missing data"}), 400

    result = collection.insert_one({
        "name": data['name'],
        "email": data.get('email', 'N/A'),
        "age": data.get('age')
    })

    return jsonify({
        "message": "User created successfully!",
        "id": str(result.inserted_id)
    }), 201

@app.route('/users', methods=['GET'])
def get_users():
    users = []
    for user in collection.find():
        user['_id'] = str(user['_id'])
        users.append(user)
    return jsonify(users), 200


seed_database()
app.run(debug=True)
```

```python
from flask import Flask, jsonify
from pymongo import MongoClient, errors
import requests
from bs4 import BeautifulSoup
import time
import threading

app = Flask(__name__)

# ======================================
# MongoDB Configuration
# ======================================
client = MongoClient("mongodb://127.0.0.1:27017/")
db = client["medical_db"]
collection = db["mayo_diseases"]

collection.create_index("Link", unique=True)

# ======================================
# Scraper Configuration
# ======================================
BASE = "https://www.mayoclinic.org"
INDEX_URL = "https://www.mayoclinic.org/diseases-conditions/index"

headers = {"User-Agent": "Mozilla/5.0"}
session = requests.Session()
session.headers.update(headers)


# ======================================
# Scraper Function
# ======================================
def scrape_and_store():
    print("\nSCRAPER STARTED\n")

    for letter in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":

        print(f"Processing letter {letter} ...")

        try:
            response = session.get(INDEX_URL, params={"letter": letter}, timeout=15)

            if response.status_code != 200:
                print("Failed to load index page")
                continue

            soup = BeautifulSoup(response.text, "html.parser")

            tags = soup.find_all("a", href=True)

            disease_links = [
                tag["href"]
                for tag in tags
                if "/diseases-conditions/" in tag["href"]
                and "/symptoms-causes/" in tag["href"]
            ]

            disease_links = list(set(disease_links))

            # -------- PRINTS YOU REQUESTED --------
            print(f"Found {len(disease_links)} diseases")
            print(f"Found {len(disease_links)} diseases for letter {letter}")

            for link in disease_links:

                full_link = BASE + link if link.startswith("/") else link

                try:
                    disease_page = session.get(full_link, timeout=15)

                    if disease_page.status_code != 200:
                        continue

                    disease_soup = BeautifulSoup(disease_page.text, "html.parser")

                    h1_tag = disease_soup.find("h1")
                    if not h1_tag:
                        continue

                    disease_name = h1_tag.get_text(strip=True)

                    wanted_sections = ["Overview", "Symptoms", "Causes", "Risk factors"]
                    section_data = {sec: "" for sec in wanted_sections}

                    for h2 in disease_soup.find_all("h2"):
                        title = h2.get_text(strip=True)

                        if title in wanted_sections:
                            content = ""
                            for sibling in h2.find_next_siblings():
                                if sibling.name == "h2":
                                    break
                                content += sibling.get_text(" ", strip=True) + " "

                            section_data[title] = content.strip()

                    document = {
                        "Disease": disease_name,
                        "Link": full_link,
                        "Overview": section_data["Overview"],
                        "Symptoms": section_data["Symptoms"],
                        "Causes": section_data["Causes"],
                        "Risk_Factors": section_data["Risk factors"]
                    }

                    try:
                        collection.insert_one(document)
                        print(f"Inserted: {disease_name}")
                        print(f"Saved: {disease_name}")

                    except errors.DuplicateKeyError:
                        print(f"Duplicate skipped: {disease_name}")
                        print(f"Saved: {disease_name}")

                    time.sleep(1)

                except Exception as e:
                    print("Error inserting disease:", e)

            print()  # blank line after each letter

        except Exception as e:
            print("Error processing letter:", e)

    print("\nSCRAPER FINISHED\n")


# ======================================
# Flask Routes
# ======================================
@app.route("/")
def home():
    return "Mayo Clinic Scraper API Running"


@app.route("/scrape")
def run_scraper():
    thread = threading.Thread(target=scrape_and_store)
    thread.start()
    return jsonify({"message": "Scraping started in background"})


@app.route("/count")
def count_documents():
    count = collection.count_documents({})
    return jsonify({"total_documents": count})


# ======================================
# Run App
# ======================================
if __name__ == "__main__":
    app.run(debug= False)
```

```python
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient
from datetime import datetime
import time

# ==============================
# MongoDB Setup
# ==============================

client = MongoClient("mongodb://127.0.0.1:27017/")
db = client["medical_db"]
collection = db["mayo_diseases"]

# Reset indexes (important if you had Link index before)
collection.drop_indexes()

# Create correct unique index
collection.create_index("url", unique=True)

# ==============================
# Scraper Config
# ==============================

BASE = "https://www.mayoclinic.org"
HEADERS = {"User-Agent": "Mozilla/5.0"}

wanted_sections = ["overview", "symptoms", "causes", "risk"]

# ==============================
# Helper function to clean text
# ==============================

def clean_text(text):
    if not text:
        return ""

    remove_words = [
        "Enlarge image Close",
        "Enlarge image",
        "Close"
    ]

    for word in remove_words:
        text = text.replace(word, "")

    return " ".join(text.split())


# ==============================
# Collect letter pages
# ==============================

print("Collecting letter links...")

main_url = f"{BASE}/diseases-conditions"
res = requests.get(main_url, headers=HEADERS)

soup = BeautifulSoup(res.text, "html.parser")

letter_links = set()

for a in soup.find_all("a", href=True):
    if "index?letter=" in a["href"]:
        letter_links.add(BASE + a["href"])

print(f"Found {len(letter_links)} letter pages")


# ==============================
# Collect disease links
# ==============================

all_disease_links = set()

for letter_url in letter_links:

    print("Scanning:", letter_url)

    try:
        res = requests.get(letter_url, headers=HEADERS)

        if res.status_code != 200:
            continue

        soup = BeautifulSoup(res.text, "html.parser")

        tags = soup.find_all(
            "a",
            class_="cmp-anchor--plain cmp-button cmp-button__link cmp-results-with-primary-name__see-link"
        )

        for tag in tags:
            href = tag.get("href")

            if href:
                if href.startswith("/"):
                    href = BASE + href

                all_disease_links.add(href)

        time.sleep(1)

    except Exception as e:
        print("Letter scan error:", e)

print("Total diseases found:", len(all_disease_links))


# ==============================
# Scrape Disease Pages
# ==============================

for url in all_disease_links:

    print("Scraping:", url)

    try:
        res = requests.get(url, headers=HEADERS)

        if res.status_code != 200:
            continue

        soup = BeautifulSoup(res.text, "html.parser")

        # Disease name
        h1 = soup.find("h1")
        if not h1:
            continue

        disease_name = clean_text(h1.get_text(strip=True))

        data = {}

        # Extract sections
        for h2 in soup.find_all("h2"):

            title = h2.get_text(strip=True).lower()

            if any(sec in title for sec in wanted_sections):

                content = ""

                for sibling in h2.find_next_siblings():

                    if sibling.name == "h2":
                        break

                    content += sibling.get_text(" ", strip=True) + " "

                data[title] = clean_text(content)

        # Save to MongoDB
        if data:

            document = {
                "disease_name": disease_name,
                "url": url,
                "sections": data,
                "scraped_at": datetime.utcnow()
            }

            collection.update_one(
                {"url": url},
                {"$set": document},
                upsert=True
            )

            print("Saved:", disease_name)

        time.sleep(0.5)

    except Exception as e:
        print("Disease scrape error:", e)

print("DONE")
```
